````python
from selenium import webdriver
driver = webdriver.Chrome()
driver.get("http://www.mineral.k12.nv.us/pages/School_Board_Minutes")
````

````python
#Dates
for i in range(5,63):
    table = '//*[@id="livesite-page-content-left"]/p['+str(i)+']/span/a/span'
    tables = driver.find_elements_by_xpath(table)
    for table in tables:
        print(table.text)
        #row['Dates'] = table.text
        row.append(table.text)
may2017 = driver.find_elements_by_xpath('//*[@id="livesite-page-content-left"]/p[62]/span/span/a/span')
for may4 in may2017:
    print(may4.text)
    row.append(may4.text)
````

````python
#URLS
for i in range(5,62):
    link = '//*[@id="livesite-page-content-left"]/p['+str(i)+']/span/a'
    links = driver.find_elements_by_xpath(link)
    for url in links:
        print(url.get_attribute('href'))
may2017 = driver.find_elements_by_xpath('//*[@id="livesite-page-content-left"]/p[62]/span/span/a')
for may4 in may2017:
    print(may4.get_attribute('href'))
````

````python
rows = []
#All code in one
may2017 = driver.find_elements_by_xpath('//*[@id="livesite-page-content-left"]/p[62]/span/span/a/span')
may2017_link = driver.find_elements_by_xpath('//*[@id="livesite-page-content-left"]/p[62]/span/span/a')

for i in range(5,62):
    table = '//*[@id="livesite-page-content-left"]/p['+str(i)+']/span/a/span'
    tables = driver.find_elements_by_xpath(table)
    link = '//*[@id="livesite-page-content-left"]/p['+str(i)+']/span/a'
    links = driver.find_elements_by_xpath(link)
    for url in links:
        for table in tables:
            row = {}
            row['Name'] = table.text
            row['Links'] = url.get_attribute('href')
            rows.append(row)
            
for may4 in may2017:
    for may4_url in may2017_link:
        row2 = {}
        row2['Name'] = may4.text
        row2['Links'] = may4_url.get_attribute('href')
        rows.append(row2)
rows
````

````python
import pandas as pd
df = pd.DataFrame(rows)
df
````

````python
df['Name'] = pd.to_datetime(df.Name)
````
