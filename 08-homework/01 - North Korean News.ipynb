```python
from selenium import webdriver
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("http://kcna.kp/")
```

```python
headlines = driver.find_elements_by_class_name("titlebet")
```

```python
for headline in headlines:
    try:
        print(headline.text)
    except:
        continue
```

```python
for a in driver.find_elements_by_class_name("titlebet"):
    print(a.get_attribute('onclick')[16:24])
```

```python
rows = []
for a in driver.find_elements_by_class_name("titlebet"):
    row = {}
    try:
        row['headline'] = a.text
    except:
        continue
    row['onclick'] = a.get_attribute('onclick')
    row['ID'] = a.get_attribute('onclick')[16:24]
    #print(row)
    rows.append(row)
```

```python
import pandas as pd
df = pd.DataFrame(rows)
df.head()
```

```python
df.to_csv('nk-news.csv', index=False)
```
