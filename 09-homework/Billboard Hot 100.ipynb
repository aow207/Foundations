Scrape the fields below, and save as a CSV file.

- Rank
- Artist name
- Song name

This one is pretty simple.

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.billboard.com/charts/hot-100"
response = requests.get(url)
doc = BeautifulSoup(response.text)
```

```python
billboard = doc.find_all(class_='chart-list-item__first-row chart-list-item__cursor-pointer')
```

```python
rows = []
for chart in billboard:
    row = {}
    row['Rank'] = chart.find(class_="chart-list-item__rank").text.strip()
    row['Artist'] = chart.find(class_="chart-list-item__artist").text.strip()
    row['Song'] = chart.find(class_="chart-list-item__title-text").text.strip()
    rows.append(row)
rows
```

```python
df = pd.DataFrame(rows)
df.head()
```

```python
df.to_csv("billboard.csv", index=False)
```
